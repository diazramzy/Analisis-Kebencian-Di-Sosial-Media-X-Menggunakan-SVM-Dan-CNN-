```python
# Load dataset
df = pd.read_csv('/content/drive/MyDrive/TA DIAZ/datasetdiaz.csv')
```

```python
# Preprocessing
df = df.dropna(subset=['Tweet', 'HS'])
X = np.array(df['Tweet'])
y = np.array(df['HS'])
```

```python
# Mengisi nilai NaN dalam X dengan string kosong
X = np.where(pd.isna(X), '', X)
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=25)
```

```python
# TF-IDF Vectorization
vectorizer = TfidfVectorizer(
    use_idf=True,
    smooth_idf=False,
    ngram_range=(1, 3),
    max_features=2000,
    min_df=20
)

X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
# Standardize features
scaler = StandardScaler(with_mean=False)
X_train_scaled = scaler.fit_transform(X_train_tfidf)
X_test_scaled = scaler.transform(X_test_tfidf)
```

```python
# Train and Evaluate SVM Models with Different Kernels
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
for kernel in kernels:
    print(f"\nTraining SVM with {kernel} kernel...")
    classifier = SVC(kernel=kernel)
    classifier.fit(X_train_scaled, y_train)

    # Predict
    y_pred = classifier.predict(X_test_scaled)

    # Evaluation
    confusion_mtx = confusion_matrix(y_test, y_pred)
    classification_rep = classification_report(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    print(f"\nConfusion Matrix ({kernel} Kernel):")
    print(confusion_mtx)
    print(f"\nClassification Report ({kernel} Kernel):")
    print(classification_rep)
    print("\nMetrics:")
    print(f"Akurasi      : {accuracy:.4f}")
    print(f"Precision    : {precision:.4f}")
    print(f"Recall       : {recall:.4f}")
    print(f"F1-Score     : {f1:.4f}")
```
